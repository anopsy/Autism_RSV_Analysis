```python
for index in languages.index:
  dates = pd.DataFrame()

  for col in languages.columns:
    lang = pd.DataFrame(languages.loc[index, col] * countries_geo[col]/100)
    dates = pd.concat([dates, lang], axis=1)

  dates.reset_index(level=1, inplace=True)
  dates = pd.melt(dates, id_vars='code', value_vars=['afrikaans', 'arabic', 'finnish', 'kroatian', 'bulgarian', 'french',
       'norwegian', 'chineeshk', 'german', 'persian', 'slovaks', 'chinees',
       'hebrew', 'spanish', 'czech', 'indonesian', 'swahili', 'dutch',
       'japanese', 'turkish', 'vietnamese', 'english', 'korean', 'polish',
       'portugese', 'russian'], var_name= 'language', value_name=index)
  geo_dates['code'] = dates['code']
  geo_dates['language'] = dates['language']
  geo_dates[index] = dates[index]
```

```python
import imageio
```

```python
def scrape_filenames():
  '''
  lists the names of uploaded files
  cleans them and returns a list of file names
  '''
  string_list = !ls
  file_list = []

  for s in string_list:
    s = s.replace('  ', ',').replace('    ', ',').replace('\t', ',')
    file_list.extend(s.split(','))

  file_list = [s for s in file_list if len(s)>0]
  file_list = [s.strip() for s in file_list]

  return file_list
```

```python
image_list = scrape_filenames()
```

```python
substr = '.png'
png_list = []
for image in image_list:
  if substr in image:
    png_list.append(image)
png_list.sort()
print(png_list)
```

```python
gif_img = []
for image in png_list:
  gif_img.append(imageio.imread(image))
```

```python
imageio.mimsave('autism.gif', gif_img, duration = 0.3)
```
